```python
import numpy as np
from lucifex.viz import plot_line, plot_twins
from formulae.flux import (concentration_sharp_front, height_theoretical, mass_capillary_trapped, 
                           mass_dissolved, fraction_capillary_trapped_sharp_front, fraction_dissolved_sharp_front)


lmbda = 0.1
sr = 0.1
h0 = 0.9
epsilon = 0.01

t = np.linspace(0, 20, 200)
theta = concentration_sharp_front(t, lmbda, Sr, h0, epsilon)
h = height_theoretical(theta, Sr, h0, epsilon)
mD = mass_dissolved(theta, Sr, h0, epsilon)
mC = mass_capillary_trapped(theta, Sr, h0, epsilon)
fD = fraction_dissolved_sharp_front(theta, Sr, h0, epsilon)
fC = fraction_capillary_trapped_sharp_front(theta, Sr, h0, epsilon)

plot_twins(t, (mD, mC), 't', ('$m_D$', '$m_C$'))
plot_line([(t, mD), (t, mC)], 't', legend_labels=('$m_D$', '$m_C$'))
plot_line([(t, fD), (t, fC)], 't', legend_labels=('$f_D$', '$f_C$'))
plot_line((t, theta), 't', '$\Theta$')
plot_line((t, h), 't', '$h$')
```
