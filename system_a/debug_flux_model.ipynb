```python
import numpy as np
from lucifex.viz import plot_line
from model.interfacial_flux import FluxModel, FluxModelEquations

Lx = 2.0
Da = 1000.0
epsilon = 0.01
h0 = 0.9
sr = 0.2
cr = 0.0

lmbda = 0.1
n = 2

t_cutoff = 1 / (epsilon * Da)
dt_early = 0.01 / Da
t_early = np.arange(0, t_cutoff, dt_early)
dt_late = 0.1
t_late = np.arange(t_cutoff, 50.0, dt_late)
t = np.concatenate((t_early, t_late))
y = np.linspace(0, 1, num=200)

model = FluxModel(t, y, Lx, Da, epsilon, h0, sr, cr, lmbda, n)
```

```python
fig, ax = plot_line((t, model.c_plus), x_label='$t$', y_label='$c^+$')
ax.set_xlim(0, 2 * t_cutoff)
ax.vlines(t_cutoff, 0, 1.0, linestyle='dashed', color='black', linewidth=0.5)

fig, ax = plot_line((t, model.c_minus), x_label='$t$', y_label='$c^-$')
fig, ax = plot_line((t, model.s_plus), x_label='$t$', y_label='$s^+$')

fig, ax = plot_line((t, model.h), x_label='$t$', y_label='$h$')

fig, ax = plot_line((t, model.f), x_label='$t$', y_label='$F$')

fig, ax = plot_line((t, model.mD), x_label='$t$', y_label='$m_D$')
fig, ax = plot_line((t, model.mC), x_label='$t$', y_label='$m_C$')
fig, ax = plot_line((t, model.mD + model.mC), x_label='$t$', y_label='$m_T$')
```
