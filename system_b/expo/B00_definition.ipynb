# Definition

$$
\mathbb{S}_{s,c,\textbf{u},p}
\begin{cases}
\Omega = \{(x,y)~:~~R_{\text{inner}}^2 < x^2 + y^2 < R_{\text{outer}}^2~,~y>0\} \\
\partial\Omega_{\text{outer}} = \{(x,y)~:~x^2 + y^2 < R_{\text{outer}}^2\} \\
\partial\Omega_{\text{inner}} = \{(x,y)~:~x^2 + y^2 < R_{\text{inner}}^2\} \\
\partial\Omega_{\text{left}} = \{(x,y)~:~y=0~,~-R_{\text{outer}}<x<-R_{\text{inner}}\} \\
\partial\Omega_{\text{right}} = \{(x,y)~:~y=0~,~R_{\text{inner}}<x<R_{\text{outer}}\} \\
(\textbf{n}\cdot\nabla\textbf{u})\vert_{\partial\Omega_{\text{outer}}}=0 & \text{no-penetration on outer boundary}\\
(\textbf{n}\cdot\nabla\textbf{u})\vert_{\partial\Omega_{\text{inner}}}=0 & \text{no-penetration on inner boundary}\\
(\textbf{n}\cdot\nabla\textbf{u})\vert_{\partial\Omega_{\text{left}}}=-PeBu/Ra & \text{inflow on left boundary}\\
(\textbf{n}\cdot\nabla\textbf{u})\vert_{\partial\Omega_{\text{right}}}=PeBu/Ra & \text{outflow on right boundary}\\
\varphi = 1 & \text{constant rock porosity} \\
\mathsf{K} = \phi^2\mathsf{I} & \text{isotropic permeability} \\
\mathsf{D} = \phi\mathsf{I} & \text{isotropic dispersion}\\
\mu = 1 & \text{constant viscosity} \\
\rho(c)=c & \text{linear density} \\
\Sigma(c, s)=s(1-c) & \text{linear reaction} \\
\implies R(s)=-s~,~J(s)=s \\
\textbf{e}_g = -\textbf{e}_y & \text{vertically downward gravity} \\ 
\end{cases}
$$
