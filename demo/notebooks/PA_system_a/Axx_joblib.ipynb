```python
from joblib import Parallel, delayed
from functools import lru_cache


# @lru_cache
def double(x):
    print('actually doing')
    return 2 * x

def job_func(x):
    import time
    import os
    time.sleep(2.0)
    print(os.getpid())
    return double(x)


Parallel(n_jobs=3)(delayed(job_func)(2) for _ in range(3)) 
```
